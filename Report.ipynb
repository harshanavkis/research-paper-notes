# Configurations and Parameters

* Model : **VGG16** with the fully connected layers included
* Split Layer : **block2_conv1** layer of VGG16
* Packet Loss Probability : $[0, 1]$ with step size of $0.05$
* Dataset : 89 images from the "screwdriver" class was used. They were downloaded in a random manner from **Google Images**
* Packet Length : Pipeline was tested on $[1, 5]$ rows per packet.
* Measured Loss : Categorical Cross Entropy

# Test 1: Basic Lossy Channel

This channel has the following properties:
   *   There is a fixed probability with which packets are lost.
   *   Currently packets that are lost are replaced by zeros at the receiver.
