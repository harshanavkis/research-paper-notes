# Test on Gilbert-Elliot channel model

## Specifications

* Loss probability: $[0.01, 0.05, 0.1]$
* Burst length : $[1, 2, 3, 4, 5]$
* Plots of accuracy vs burst length for each loss probability

```python
import numpy as np
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
dir  = 'gilbert'
```

```python
lossList = np.array([0.01, 0.05, 0.01])
burstLength = np.arange(1, 6, 1)
```

## Layer: block3_pool

```python
layerDir = os.path.join(dir, 'block3_pool')
testDirs = os.listdir(layerDir)[:-1]
testDirs = [os.path.join(layerDir, i) for i in testDirs]
testDirs
```

```python
loss0Dir = testDirs[0]
loss0Dir = os.listdir(loss0Dir)
loss0Dir = [os.path.join(testDirs[0], i) for i in loss0Dir]

loss1Dir = testDirs[1]
loss1Dir = os.listdir(loss1Dir)
loss1Dir = [os.path.join(testDirs[1], i) for i in loss1Dir]

loss2Dir = testDirs[2]
loss2Dir = os.listdir(loss2Dir)
loss2Dir = [os.path.join(testDirs[2], i) for i in loss2Dir]
```

```python
def calcMean(fileName):
    a = np.load(fileName)
    a = np.mean(a, axis=0)
    return a[1]
```

```python
loss0List = np.array([calcMean(i) for i in loss0Dir])
print(loss0List)
loss1List = np.array([calcMean(i) for i in loss1Dir])
print(loss1List)
loss2List = np.array([calcMean(i) for i in loss2Dir])
print(loss2List)
```

```python
plt.plot(burstLength, loss0List, label='1 percent')
plt.plot(burstLength, loss1List, label='5 percent')
plt.plot(burstLength, loss2List, label='10 percent')
plt.xlabel('Burst Length')
plt.ylabel('Classification Accuracy')
plt.legend(loc='upper right')
plt.show()
```
